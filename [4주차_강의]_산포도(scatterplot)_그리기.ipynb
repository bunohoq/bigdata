```python
import matplotlib.pyplot as plt
plt.title('legend') #제목넣기
plt.plot([10,20,30,40], label='asc')
plt.plot([40,30,20,10], label='desc')
plt.legend(loc=10) #범례위치
plt.show()
```

```python
#색상 바꾸기

#색깔명은 https://matplotlib.org/stable/tutorials/colors/colors.html 에서 골라보자

import matplotlib.pyplot as plt
plt.title('color') 
plt.plot([10,20,30,40], color='skyblue', label='asc')
plt.plot([40,30,20,10], 'tomato', label='desc')
plt.legend(loc=4) 
plt.show()
```

```python
# 선모양은 https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle 에서 골라보자.

import matplotlib.pyplot as plt
plt.title('linestyle') 
plt.plot([10,20,30,40], color='skyblue', linestyle='--', label='dashed')
plt.plot([40,30,20,10], 'tomato', ls=':', label='dotted')
plt.legend(loc=4) 
plt.show()
```

```python
#마커모양 바꾸기

import matplotlib.pyplot as plt
plt.title('marker') 
plt.plot([10,20,30,40], 'r.', label='circle')
plt.plot([40,30,20,10], 'g^', label='triangle')
plt.legend(loc=4) 
plt.show()
```

# **산포도(산점도, scatterplot)**
*   두 변수(x와 y)의 상관관계를 표현할 때 사용함


```python
#임의의 데이터를 생성하여 예제를 그려보자
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4], [10,30,20,40]) #x 와 y에 각각 값 입력
plt.show()
```


# 각 점의 크기를 표현하는 법 (예: 여러 국가의 산포도 그릴 때 인구 수 표현): 옵션 s 


```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4], [10,30,20,40],s=[100,200,250,300])
plt.show()
```


# 색깔 변경하는 법: 옵션 c

*   색깔명은 https://matplotlib.org/stable/tutorials/colors/colors.html 에서 골라보자

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3,4], [10,30,20,40],s=[100,200,250,300], c=['red','blue', 'lavender', 'gold'])
plt.show()
```
